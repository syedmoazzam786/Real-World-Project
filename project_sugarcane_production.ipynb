```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
import warnings
```

```python
df = pd.read_csv("/content/drive/MyDrive/dataset/List of Countries by Sugarcane Production.csv")
```

```python
df.head()
```

```python
df.shape
```

## Data cleaning

```python


# If there's a typo, correct it in the code below
df["Production (Tons)"] = df["Production (Tons)"].str.replace(".", "")
```

```python

# Use the correct column name and remove any extra spaces or tabs.
df["Production per Person (Kg)"] = df["Production per Person (Kg)"].str.replace(".", "").str.replace(",", ".")
```

```python
df["Acreage (Hectare)"] = df["Acreage (Hectare)"].str.replace(".", "")
```

```python
df["Yield (Kg / Hectare)"] = df["Yield (Kg / Hectare)"].str.replace(".", "").str.replace(",", ".")
```

```python
df.head()
```

```python
## rename and remove gaps in columns
df.rename(columns={"Production (Tons)": "Production(tons)"}, inplace=True)
```

```python
df.rename(columns={"Acreage (Hectare)": "Acreage(hectare)"}, inplace=True)
```

```python
df.rename(columns={"Production per Person (Kg)": "Productionperperson(kg)"}, inplace=True)
```

```python
df.rename(columns={"Yield (Kg / Hectare)": "Yield(kg/hectare)"}, inplace=True)
```

```python
df.head()
```

```python
#check the null values
df.isna().sum()
```

```python
df[df["Acreage(hectare)"].isnull()]
```

```python
 df = df.dropna().reset_index()
```

```python
df.head()
```

```python
df.drop(["index","Unnamed: 0"], axis=1, inplace=True)
```

```python
df.head()

```

```python
df.dtypes
```

```python
df["Production(tons)"] = df["Production(tons)"].astype(float)
```

```python
df["Productionperperson(kg)"] = df["Productionperperson(kg)"].astype(float)
```

```python
df["Acreage(hectare)"] = df["Acreage(hectare)"].astype(float)
```

```python
df["Yield(kg/hectare)"] = df["Yield(kg/hectare)"].astype(float)
```
